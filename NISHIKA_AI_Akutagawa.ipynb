```python
# trainデータをgoogle driveから読み込み
import pandas as pd
train_df = pd.read_csv('/content/drive/MyDrive/train.csv')
```

```python
train_df.info()
```

```python
train_df.head(20)
```

```python
# trainデータから不要な列（writing_id）を削除し、特徴量（X）と目的変数（y）に仕分け
X = train_df.drop(['writing_id', 'author'], axis=1)
y = train_df['author']
```

```python
# TF-IDFを用いて自然言語処理
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_body = vectorizer.fit_transform(X['body'])
```

```python
# 自然言語処理済みの特徴量（X_body）を使って訓練用データと検証用データを分割
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X_body, y, test_size=0.3, random_state=0)
```

```python
# ロジスティクス回帰モデルで訓練
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# F値（F1Score）を用いて精度を検証
from sklearn.metrics import f1_score
y_pred = model.predict(X_valid)
f1_score(y_valid, y_pred, average='macro')
```

```python
# testデータを読み込む
test_df = pd.read_csv('/content/drive/MyDrive/test.csv')
```

```python
# testデータの内容を確認
test_df.info()
test_df.head(10)
```

```python
# writng_idを項目削除（trainデータと同じ処理）
test_df_body = test_df.drop(['writing_id'], axis=1)
```

```python
# test_df_authorの項目を確認
test_df_body.head(10)
```

```python
# test_df_bodyをTF-IDFを使って自然言語処理
test_body = vectorizer.transform(test_df_body['body'])
```

```python
# testデータを使って学習済モデルでauthor（芥川龍之介ornot）を判別
y_pred_author = model.predict(test_body)
```

```python
# 提出用データフレームを作成
submission = pd.DataFrame({
    'writing_id':test_df['writing_id'],
    'author':y_pred_author})

# csvとして保存
submission.to_csv('/content/drive/MyDrive/submission_1.csv', index=False)
```
